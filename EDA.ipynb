```python
# user defined function to check the detailed info of dataframe

def dataframeinfo(df):
    data_types = df.dtypes.unique()

    max_length = max([len(col) for col in df.columns])
    if max_length > 13:
        max_length = max_length - 13
    else:
        max_length = 1
    print("Total rows: ", df.shape[0])
    print("Total columns: ", df.shape[1])
    print('')
    print(f"#      Column Name      {' ' * max_length} Non Null Count      Null Count      Non Null Percentage      Null Percentage      Data Type      Unique Values")
    print(f"--     -----------      {' ' * max_length} --------------      ----------      -------------------      ---------------      ---------      -------------")

    for i, value in enumerate(df.columns.tolist()):
        null_count = df[value].isnull().sum()
        non_null_Count = df.shape[0]-null_count
        non_null_percentage = (df.shape[0] - null_count) / df.shape[0]
        null_percentage = 1 - non_null_percentage
        data_type = str(df[value].dtype)
        unique_values = df[value].nunique()
        print(f'{i: <6} {value: <{17+max_length}} {non_null_Count: <19} {null_count: <15} {non_null_percentage: <24.2%} {null_percentage: <20.2%} {data_type: <14} {unique_values}')

    print("")
    print('--------------------------')
    print("Datatype Count:")
    for dtype in data_types:
        dtype_count = len(df.select_dtypes(include=[dtype]).columns)
        print(f" - {dtype}: {dtype_count}")
```

```python
# Let us drop unique_id column as it does not contribute to the model

df.drop(columns=['unique_id'], inplace=True)
```

```python
# fill empty in object types with mode and numerical types with median

for column in df.columns:
    if df[column].dtype == 'object':
        df[column] = df[column].fillna(df[column].mode()[0])
    else:
        df[column] = df[column].fillna(df[column].median())
```

```python
# convert float64 to int 64

float_columns = df.select_dtypes(include='Float64').columns
df[float_columns] = df[float_columns].astype('int64')
```

```python
def feature_vs_target_bivariate(df, target):
    features = [col for col in df.columns if col != target]
    num_features = len(features)
    num_rows = (num_features + 2) // 3  # Calculate the number of rows needed

    for i in range(num_rows):
        fig, axes = plt.subplots(1, 3, figsize=(24, 6))
        
        for j in range(3):
            index = i * 3 + j
            if index < num_features:
                feature = features[index]
                crosstab = pd.crosstab(df[feature], df[target])
                crosstab.plot(kind='bar', stacked=True, ax=axes[j])
                axes[j].set_title(f'Stacked Bar Plot between {feature} and {target}')
                axes[j].set_xlabel(feature)
                axes[j].set_ylabel('Frequency')
            else:
                axes[j].axis('off')  # Turn off axis if there are no more features to plot
        
        plt.tight_layout()
        plt.show()


feature_vs_target_bivariate(df, 'h1n1_vaccine')
```

```python
def plot_pie_charts(df, columns, colors=None):
    num_columns = 3
    num_plots = len(columns)
    num_rows = (num_plots + num_columns - 1) // num_columns
    
    fig, axes = plt.subplots(num_rows, num_columns, figsize=(21, 7 * num_rows))
    axes = axes.flatten()

    for i, column in enumerate(columns):
        # Get the value counts for the column and plot as a pie chart
        df[column].value_counts().plot(
            kind='pie',
            autopct='%0.2f%%',
            explode=[0.1 if j == 0 else 0 for j in range(len(df[column].value_counts()))],
            ax=axes[i],
            colors=colors
        )
        axes[i].set_title(f'Distribution of {column}')
        axes[i].set_ylabel('')

    # Hide any unused subplots
    for j in range(i + 1, len(axes)):
        fig.delaxes(axes[j])
        
    plt.tight_layout()
    plt.show()

# List of all columns to plot pie charts for
columns = df.columns

# Define a color palette if desired
colors = ['#1679AB', '#3C5B6F', '#948979', '#DFD0B8', '#1679AB']

# Plot pie charts for all columns
plot_pie_charts(df, columns, colors)
```

```python
# Label Encoding Categorical data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()


df_e = df.copy()
object_columns = df_e.select_dtypes(include='object').columns

for col in object_columns:
  print('')
  print(f'Encoded value of column {col}')
  if df_e[col].dtype == 'object':
    df_e[col] = df_e[col].str.strip().str.title()
  df_e[col] = le.fit_transform(df_e[col])
  print(dict(zip(le.classes_, range(len(le.classes_)))), end= '')
  print('')
```

```python
def plot_heatmap(df):
    plt.figure(figsize=(30, 30))
    sns.heatmap(df.corr(), annot=True, fmt="0.2f", cmap='coolwarm', cbar=True, square=True, annot_kws={"size": 10})
    plt.title('Correlation Heatmap', fontsize=20)
    plt.xticks(rotation=45, ha='right', fontsize=10)
    plt.yticks(fontsize=10)
    plt.show()

# Plot the heatmap
plot_heatmap(df_e)
```

```python
def display_correlation_matrix(df):
    corr_matrix = df.corr()
    print("Correlation Matrix:")
    for i in range(len(corr_matrix.columns)):
        for j in range(i + 1, len(corr_matrix.columns)):
            if corr_matrix.iloc[i, j] > 0.5 or corr_matrix.iloc[i, j] < -0.5:
                print(f"{corr_matrix.index[i]} - {corr_matrix.columns[j]}: {corr_matrix.iloc[i, j]}")

display_correlation_matrix(df_e)
```
